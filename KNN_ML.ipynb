```python
'''
X = data.iloc[:,0:30].values
y = data.iloc[:,31].values


from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.30)  

from sklearn.preprocessing import StandardScaler    
st_X= StandardScaler()    
X_train= st_X.fit_transform(X_train)    
X_test= st_X.transform(X_test)  
'''


x1 = data.iloc[:,0:30].values
y1 = data.iloc[:,31].values
x1=pd.DataFrame(x1)
from sklearn.model_selection import train_test_split  
x1_train, x1_test, y1_train, y1_test= train_test_split(x1, y1, test_size= 0.30, random_state=0)  

from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x1_train= st_x.fit_transform(x1_train)    
x1_test= st_x.transform(x1_test) 
```

```python
# import KNeighbors ClaSSifier from sklearn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x1_train, y1_train)
```

```python
y1_pred = knn.predict(x1_test)
y1_pred
```

```python
# probability of getting output as 2 - benign cancer
knn.predict_proba(x1_test)[:,0]
```

```python
# probability of getting output as 4 - malignant cancer
knn.predict_proba(x1_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y1_test, y1_pred)))
```

Compare the train-set and test-set accuracy
Now, I will compare the train-set and test-set accuracy to check for overfitting.

```python
y_pred_train = knn.predict(x1_train)
y_pred_train

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y1_train, y_pred_train)))
```

Check for overfitting and underfitting¶


```python
print('Training set score: {:.4f}'.format(knn.score(x1_train, y1_train)))

print('Test set score: {:.4f}'.format(knn.score(x1_test, y1_test)))
```

The training-set accuracy score is 0.9821 while the test-set accuracy to be 0.9714. These two values are quite comparable. So, there is no question of overfitting.

```python
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(x1_train, y1_train)
y_pred_5 = knn_5.predict(x1_test)
print('Model accuracy score with k=5 : {0:0.4f}'. format(accuracy_score(y1_test, y_pred_5)))
```
